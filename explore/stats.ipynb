```python
import pandas as pd
from pathlib import Path
import h5py
import tqdm
import plotly.express as px

root = Path("../data/")
dataset = root / "input" / "trat3_production_1650_1700_20231411_v1.hdf5"
```

```python
L = []
with h5py.File(dataset, "r") as fp:
    for v in tqdm.tqdm(fp.values(),total=len(fp)):
        L.append({**v.attrs})
```

```python
df = pd.DataFrame(L)

print(df.columns)
display(df.head())
```

```python
# Count the number of samples by log_name
df["build_log"] = df["planid"] + ", " + df["log_name"]
df.to_json("../data/stats_dataset.json")
counts_by_log_plan = df.groupby("name").size().reset_index(name='count')
display(counts_by_log_plan)
# Plot histogram using plotly express
fig = px.histogram(counts_by_log_plan, x="count", labels={'count': 'Number of events/lines'}, cumulative=True)
fig.update_layout(
    yaxis_title_text='Number of files',
)
# Show the plot
fig.show()
```
